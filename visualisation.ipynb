```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import json
```

```python
with open('statistic.txt', 'r') as f:
    dict_ = json.loads(f.read())
table = pd.DataFrame(dict_)
```

```python
#define plot size in inches (width, height) & resolution(DPI)
fig = plt.figure(figsize=(12, 9), dpi=200)
plt.plot(range(table.shape[0]), table['queue_size_per_second'], color='r', label='queue_size_per_second')
plt.plot(range(table.shape[0]), table['amount_task_per_second'], color='b', label='amount_tasks_per_second')
plt.title('Consumer/Producer activity', fontsize=20)
plt.legend(loc='upper right', prop={'size': 10})
plt.xlabel('seconds')
plt.ylabel('amount')
plt.yticks(np.arange(min(table['queue_size_per_second']), max(table['queue_size_per_second'])+1, 1.0))
plt.grid()
plt.show() 
```
